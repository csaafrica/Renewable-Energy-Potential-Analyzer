```python

def get_nasa_monthly_avg_weather(town_name, latitude, longitude, start_year=2018, end_year=2024):
    """
    Fetches NASA POWER monthly average temperature and wind speed for a given location.

    Parameters:
    - town_name (str): Name of the town
    - latitude (float)
    - longitude (float)
    - start_year (int)
    - end_year (int)

    Returns:
    - pd.DataFrame with columns: Town, Year, Month, Avg_Temperature_C, Avg_Wind_Speed_m/s
    """
    url = (
        f"https://power.larc.nasa.gov/api/temporal/monthly/point"
        f"?parameters=T10M,WS50M,PS"
        f"&community=RE"
        f"&latitude={latitude}"
        f"&longitude={longitude}"
        f"&start={start_year}"
        f"&end={end_year}"
        f"&format=JSON"
    )

    try:
        response = requests.get(url)
        response.raise_for_status()#raise an error
        data = response.json()

        temp_data = data['properties']['parameter']['T10M']#for getting the temperature at 50M above the surface
        wind_data = data['properties']['parameter']['WS50M']#for getting the wind speed at 50M above the surface
        pressure_data = data['properties']['parameter']['PS'] 

        rows = []
        for date_str in temp_data:
            year = int(date_str[:4])#since date is of the formart 20191 weare taking the first 4 values for year and the remainder for month
            month = int(date_str[4:])
            rows.append({
                "Town": town_name,
                "Year": year,
                "Month": month,
                "Avg_Temperature_C": temp_data[date_str],
                "Avg_Wind_Speed_m/s": wind_data.get(date_str, None),
                "Surface_Pressure_kPa": pressure_data.get(date_str)
            })#we are now adding the dictionary we created here into our rows list we are using .get() to handle missing wind/Pa data

        return pd.DataFrame(rows)

    except Exception as e:
        print(f"Failed to fetch data for {town_name}: {e}")
        return pd.DataFrame()
```

```python
all_nasa_weather = []

for town in latitude_dict:
    lat = latitude_dict[town]
    lon = longitude_dict[town]

    df = get_nasa_monthly_avg_weather(town_name=town, latitude=lat, longitude=lon, start_year=2018, end_year=2024)
    if not df.empty:
        all_nasa_weather.append(df)

nasa_final_df = pd.concat(all_nasa_weather, ignore_index=True)
nasa_final_df.to_excel('nasa_dataset_50M.xlsx', index=False)
```
